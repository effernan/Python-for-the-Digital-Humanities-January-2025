#### A. Import the libraries

```python
#Your code in here
```

```python
#Solution
from urllib import request
from bs4 import BeautifulSoup
import re
import pandas as pd
```

#### B. Get the data from the internet

```python
#Your code in here
```

```python
#Solution
url = "https://www.gutenberg.org/cache/epub/54873/pg54873.txt"
response = request.urlopen(url)
raw = response.read().decode("utf-8")
```

#### C. Prettify the data

```python
#Your code in here
```

```python
#Solution
soup = BeautifulSoup(raw, "html.parser")
data = soup.prettify()
```

#### D. Select the data

```python
#Solution
text = re.search("START OF THE PROJECT GUTENBERG EBOOK", data)
text
```

# Exercise 3: creating a Data Frame

#### A. Create a list of chapters

```python
#your code here
```

```python
#Solution

chapter = []

x = list(range(1, 48)) #we need to write 38 due to Python notation
for i in x:
    chapter.append(f"Chapter {i}")
print(chapter)
```

#### B. Create a dictionary with the text of the book and your list of chapters

```python
#your code here
```

```python
#solution

key_list = chapter
value_list = chapters
```

```python
data = dict(zip(key_list, value_list))
```

#### C. Trasform that into a Pandas dataframe

```python
#your code in here
```

```python
#solution

data = pd.DataFrame(chapter, columns = ["Chapter"]) #so the only thing that we need to do is to create a dataframe with pandas.
data["Text"] = chapters
```

#### D. Save it into your computer as a CSV file

```python
data.to_csv("vingt_mille_chapters.csv")
```
